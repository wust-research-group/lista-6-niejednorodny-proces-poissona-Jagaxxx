#### ZADANIE 1 - METODA PRZERZEDZANIA
Napisz algorytm, który dla ustalonej niejednorodnej funkcji intensywności $\lambda(t)$ generuje czasy oczekiwania niejednorodnego procesu Poissona metodą przerzedzania.

Sprawdź poprawność jego działania weryfikując, że: $N_t$ ~ Poiss $(\int_0 ^t\lambda(s)ds)$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lambda_func(x):
    return x**2


def thinning(T):
    t = 0
    I = 0
    S = list()
    lambda_ = np.max(lambda_func(np.linspace(0, T, 1000)))
    while True:
        U1 = np.random.uniform(0, 1)
        t = t - 1 / lambda_ * np.log(U1)
        if t > T:
            break
        U2 = np.random.uniform(0, 1)
        if U2 <= lambda_func(t) / lambda_:
            I += 1
            S.append(t)
    return S
```

```python
result_thinning = thinning(T=5)

plt.figure(figsize=(8, 5))
plt.step(result_thinning, range(len(result_thinning)))
plt.xlabel("Czas")
plt.ylabel("Liczba zdarzeń")
plt.grid(True)
plt.show()
```
